# Distribution plots for numeric columns

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('heart.csv')
df.head()
    
```

```python
sns.distplot(df['Age'],kde=True,hist=True,rug= True,bins=20)
plt.show()
```

```python
sns.distplot(df['RestingBP'],kde= True,color='blue')
plt.show()
```

```python
sns.distplot(df['Cholesterol'],kde= True,color='blue')
plt.show()
```

```python
sns.distplot(df['FastingBS'],kde= True,color='blue')
plt.show()
```

```python
sns.distplot(df['MaxHR'],kde= True,color='blue')
plt.show()
```

```python
sns.distplot(df['Oldpeak'],kde= True,color='blue')
plt.show()
```

# Pie chart for categorical features

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct="%.1f")
plt.show()
```

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct="%.1f")
plt.show()
```

```python
df.groupby('ST_Slope').size().plot(kind='pie',autopct="%.0f")
plt.show()
```

```python
df.groupby('HeartDisease').size().plot(kind='pie',autopct="%.0f")
plt.show()
```
